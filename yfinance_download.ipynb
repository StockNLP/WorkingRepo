```python
# Import package
import yfinance as yf
```

```python
get_tick = ['AMC', 'GME', 'TSLA', 'MSFT', 'AAPL', 'NKLA', 'GOOGL','RIVN', 'FB']

# Get the data
for ticker in get_tick:
    data = yf.download(tickers=ticker, period="1mo", interval="15m")
    data.to_csv(ticker+'.csv')

```
