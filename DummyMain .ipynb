```python
get_tick = ['$AMC',  '$GMC', 
            '$TSLA', '$MSFT', 
            '$AAPL', '$DOGE', 
            '$SHIB', '$ETH', '$PROG', 
            '$BTC', '$ADA', 
            '$NKLA', '$GOOGL',
            '$RIVN', '$FB']
df_t = pd.DataFrame(columns=['Tweet','Tick', 'Source','Date','Likes', 'Retweet Count'])
```

```python
def get_stock_tweets (search_stock, start_date, tweet_count):
  
        new_search = search_stock + " -filter:retweets"
        tweets = tw.Cursor(api.search_tweets,
                            q=new_search,
                            lang="en",
                            since=start_date,
                            result_type = "recent").items(tweet_count)
        tweets_list = [[tweet.text, tweet.id,
                        tweet.created_at.replace(tzinfo=datetime.timezone.utc).strftime("%m/%d/%Y"),
                    tweet.favorite_count,tweet.retweet_count]for tweet in tweets] 
        return tweets_list
    
def get_tweets():
    for i in range(len(get_tick)):
        ls = get_stock_tweets(get_tick[i], date.today(), 1)
        tck_lst,src = td.tick_list(get_tick[i], ls)
        date_list,like_list, rtweet_list = td.get_like_retweet_lst(ls)
        td.add_to_df(ls,df_t, tck_lst, src, date_list, like_list, rtweet_list)
```
